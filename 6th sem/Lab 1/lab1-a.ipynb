```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoid Function

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
data = [1.0,-10.0,0.0,15.0,-2.0]
for num in data:
    print("Applying Sigmoid Activation Function on (%.1f) gives %.1f" %(num,sigmoid(num)))
```

```python
x = np.linspace(-10,10,50)
p = sigmoid(x)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.plot(x,p)
plt.show()
```

RELU

```python
def relu(x):
    return np.maximum(0, x)
```

```python
for num in data:
    print("Applying RELU Activation Function on (%.1f) gives %.1f" %(num,relu(num)))
```

```python
x = np.linspace(-10, 10, 50)
p = relu(x)
plt.plot(x, p)
plt.xlabel("x")
plt.ylabel("RELU(x)")
plt.title("RELU Function")
plt.show()
```

Leaky RELU

```python
def leaky_relu(x): 
    return np.where(x > 0, x, x * 0.01)
```

```python
for num in data:
    print("Applying Leaky RELU Activation Function on (%.1f) gives %.1f" %(num,leaky_relu(num)))
```

```python
x = np.linspace(-10,10,50)
p = leaky_relu(x)
plt.xlabel("x")
plt.ylabel("Leaky_RELU(x)")
plt.title("Leaky RELU Function")
plt.plot(x,p)
plt.show()
```

Tanh

```python
def tanh(x):
    return np.tanh(x)
```

```python
for num in data:
    print("Applying Tanh Activation Function on (%.1f) gives %.1f" %(num,leaky_relu(num)))
```

```python
x = np.linspace(-10,10,50)
p = tanh(x)
plt.xlabel("x")
plt.ylabel("Tanh(x)")
plt.title("Tanh Function")
plt.plot(x,p)
plt.show()
```

Softmax

```python
def softmax(x):
    exp_score = np.exp(x)
    return exp_score/np.sum(exp_score)
```

```python
for num in data:
    print("Applying Softmax Activation Function on (%.1f) gives %.1f" %(num,leaky_relu(num)))
```

```python
x = np.linspace(-10,10,50)
p = softmax(x)
plt.xlabel("x")
plt.ylabel("Softmax(x)")
plt.title("Softmax Function")
plt.plot(x,p)
plt.show()
```

Binary Step

```python
def binary_step(x,t=0):
    return np.heaviside(x-t,0)
```

```python
for num in data:
    print("Applying Binary Step Activation Function on (%.1f) gives %.1f" %(num,binary_step(num)))
```

```python
x = np.linspace(-10,10,50)
p = binary_step(x)
plt.xlabel("x")
plt.ylabel("Binary_Step(x)")
plt.title("Binary Step Function")
plt.plot(x,p)
plt.show()
```
