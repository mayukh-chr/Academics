```python
import numpy as np
import matplotlib.pyplot as plt

def sig(x):	
	return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 50)
p = sig(x)
plt.figure(figsize=(10, 6)) # Adjust figure size if needed
plt.plot(x, p)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.title("Sigmoid Activation Function")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
    return np.maximum(0, x)
# Generate input values (e.g., from -5 to 5)
x = np.linspace(-5, 5, 100)
# Calculate ReLU values
y = relu(x)
# Create the plot
plt.figure(figsize=(8, 6)) # Optional: Adjust figure size
plt.plot(x, y, label='ReLU')
plt.xlabel('Input (x)')
plt.ylabel('Output (ReLU(x))')
plt.title('ReLU Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def leaky_relu(x):
    if x > 0:
        return x
    else:
        return 0.1 * x # Note: 0.1 is a common value for the alpha
x = 1.0
print('Applying Leaky ReLU on (%.1f) gives %.1f' % (x, leaky_relu(x)))
x = -10.0
print('Applying Leaky ReLU on (%.1f) gives %.1f' % (x, leaky_relu(x)))
x = 0.0
print('Applying Leaky ReLU on (%.1f) gives %.1f' % (x, leaky_relu(x)))
x = 15.0
print('Applying Leaky ReLU on (%.1f) gives %.1f' % (x, leaky_relu(x)))
```

```python
def tanh(x):
    return np.tanh(x)
in_array = np.linspace(np.pi, -np.pi, 13)
out_array = np.tanh(in_array)
print("in_array:", in_array)
print("\nout_array:", out_array)
# red for numpy.tanh()
plt.plot(in_array, out_array, color="red", marker="o")
plt.title("numpy.tanh()")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def binary_step(x):
    """
    Calculates the binary step activation function.
    Args:
    x: Input value (can be a scalar or an array).
    Returns:
    1 if x >= 0, 0 otherwise.
    """
    return np.where(x >= 0, 1, 0)
# Generate input values (e.g., from -5 to 5)
x = np.linspace(-5, 5, 100)
# Calculate binary step values
y = binary_step(x)
# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Binary Step')
plt.xlabel('Input (x)')
plt.ylabel('Output (Binary Step(x))')
plt.title('Binary Step Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
    """
    Calculates the linear activation function.
    Args:
    x: Input value (can be a scalar or an array).
    Returns:
    The input value itself.
    """
    return x
# Generate input values (e.g., from -5 to 5)
x = np.linspace(-5, 5, 100)
# Calculate linear values
y = linear(x)
# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Linear')
plt.xlabel('Input (x)')
plt.ylabel('Output (Linear(x))')
plt.title('Linear Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```
